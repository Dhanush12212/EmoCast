```python
from keras.models import load_model
import cv2
import numpy as np
from keras.preprocessing.image import img_to_array

# Load the full model (architecture + weights)
model = load_model("model.h5")
model.summary()   # ✅ you already saw this output

# Define your emotion labels (5 classes since Dense layer = 5)
class_labels = ['Angry','Happy','Neutral','Sad','Surprise']


# Load face detector
faceface_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
print("Cascade loaded:", not face_classifier.empty())
_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Start video
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    print("Faces detected:", len(faces))

    for (x, y, w, h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_gray = cv2.resize(roi_gray,(48,48), interpolation=cv2.INTER_AREA)

        if np.sum(roi_gray) != 0:
            # Convert depending on model input
            if model.input_shape[-1] == 1:   # expects grayscale
                roi = roi_gray.astype("float") / 255.0
                roi = np.expand_dims(roi, axis=-1)  # add channel dim → (48,48,1)
                roi = np.expand_dims(roi, axis=0)
        
            roi = np.expand_dims(roi, axis=0)

            preds = model.predict(roi)[0]
            print("Prediction probs:", preds)  # debug
            label = class_labels[preds.argmax()]
            cv2.putText(frame, label, (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

    cv2.imshow('Emotion Detector', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
